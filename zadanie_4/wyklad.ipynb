```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import AgglomerativeClustering

# Generowanie danych "nosy circles"
X, _ = make_circles(n_samples=500, noise=0.1, factor=0.5, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], cmap='viridis', edgecolors='k')
plt.title('Nosy Circles')
plt.xlabel('Wymiar 1')
plt.ylabel('Wymiar 2')
plt.show()
```

```python

from sklearn.cluster import AgglomerativeClustering
# Wykonanie grupowania hierarchicznego z uyciem AgglomerativeClustering
n_clusters = 4
clustering = AgglomerativeClustering(n_clusters=n_clusters)
clusters = clustering.fit_predict(X)

# Wizualizacja wynikw grupowania
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', edgecolors='k')
plt.title('Nosy Circles z grupowaniem aglomeracyjnym')
plt.xlabel('Wymiar 1')
plt.ylabel('Wymiar 2')
plt.show()


```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Generowanie przykadowych danych
data = X

# Obliczanie macierzy pocze (linkage matrix)
Z = linkage(data, method='ward')

# Narysowanie dendrogramu
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram dla klasteryzacji hierarchicznej')
plt.xlabel('Prbki')
plt.ylabel('Odlego')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler  = StandardScaler().fit(X)
X_scaled = standard_scaler.transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering

n_clusters = 5
clustering = AgglomerativeClustering(n_clusters=n_clusters)
clusters = clustering.fit_predict(X_scaled)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled['cluster'] = clusters
```

```python

# Wizualizacja wynikw grupowania
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled['area_m2'], X_scaled['price_total_zl_cleaned'], c=X_scaled['cluster'], cmap='viridis', edgecolors='k')
plt.title('Nosy Circles z grupowaniem aglomeracyjnym')
plt.xlabel('area_m2')
plt.ylabel('price_total_zl_cleaned')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Generowanie przykadowych danych
data = X_scaled

# Obliczanie macierzy pocze (linkage matrix)
Z = linkage(data, method='ward')

# Narysowanie dendrogramu
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram dla klasteryzacji hierarchicznej')
plt.xlabel('Prbki')
plt.ylabel('Odlego')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Wykonanie grupowania hierarchicznego z uyciem Kmeans
n_clusters = 6
clustering = KMeans(n_clusters=n_clusters, random_state=0, n_init="auto")
clusters = clustering.fit_predict(X_scaled)

# Wizualizacja wynikw grupowania
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled['area_m2'], X_scaled['price_total_zl_cleaned'], c=clusters, cmap='viridis', edgecolors='k')
plt.scatter(clustering.cluster_centers_[:, 0], clustering.cluster_centers_[:, 1], s=100, marker='X', c='red')
plt.title('area_m2 vs price_total_zl_cleaned with KMeans Clustering')
plt.xlabel('area_m2')
plt.ylabel('price_total_zl_cleaned')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Wykonanie grupowania hierarchicznego z uyciem Kmeans
n_clusters = 6
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init="auto")
kmeans_clusters = kmeans.fit_predict(X)

# Wizualizacja wynikw grupowania
plt.figure(figsize=(8, 6))
plt.scatter(X['area_m2'], X['price_total_zl_cleaned'], c=kmeans_clusters, cmap='viridis', edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, marker='X', c='red')
plt.title('non standardized area_m2 vs price_total_zl_cleaned with KMeans Clustering')
plt.xlabel('area_m2')
plt.ylabel('price_total_zl_cleaned')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
results = []
for k in range(1, 11):
 kmeans = KMeans(n_clusters=k, random_state=0)
 kmeans.fit(X_scaled)
 results.append(kmeans.inertia_)
 if k > 1:
    silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
    print(f'For n_clusters = {k}, Silhouette Score: {silhouette_avg}')

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), results, marker='o')
plt.xlabel('Liczba klastrw')
plt.ylabel('Wariancja wewntrz klastrw')
plt.title('Wybr liczby klastrw')
plt.show()

```
